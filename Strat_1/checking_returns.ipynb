```python
import os 
import pandas as pd 
import numpy as np
os.getcwd() 
```

```python
files = [i for i in os.listdir('strat_returns') if i.endswith('csv')]
ticker = files.pop(0)
print('Ticker: ', ticker)
print('Files: ', files)
```

```python
df_hk = pd.read_csv(f'strat_returns/Testing/kelly_returns/{ticker}')
df_hk['ticker'] = ticker.replace('.csv', '')
df_hk
```

```python
df_ret = pd.read_csv(f'strat_returns/Testing/no_kelly_criterion_returns/{ticker}')
df_ret['ticker'] = ticker.replace('.csv', '')
#df_ret = df_ret[['ticker', 'date', 'daily_ret']]
df_ret[df_ret['date'] >= '2012-02-09']
```

```python
for file in files: 
    
    df = pd.read_csv(f'strat_returns/Testing/kelly_returns/{file}')
    df['ticker'] = file.replace('.csv', '')
    df_hk = pd.concat([df_hk, df], axis = 0)
    
    
df_hk = df_hk[['ticker', 'date', 'daily_ret']]
df_hk.head()
```

```python
for file in files: 
    
    df = pd.read_csv(f'strat_returns/Testing/no_kelly_criterion_returns/{file}')
    df['ticker'] = file.replace('.csv', '')
    df_ret = pd.concat([df_ret, df], axis = 0)
    
df_ret = df_ret[['ticker', 'date', 'daily_ret']]

df_ret.head()
```

```python
df1 = df_hk.merge(df_ret, how='inner', on=['ticker','date'], suffixes=['_hk', '_standard'])
df1
```
